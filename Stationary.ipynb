```python
#To check where timeseries dataset is stationary or not
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
temp=pd.read_csv(r"C:\Users\Asus\Downloads\archive\Electric_Production.csv",index_col=0) #importing dataset and marking date field as index

temp.head()
```

```python
temp["IPG2211A2N"].plot(ylim=[40,150],title='Electric production rate')
```

```python
temp['12-month-SMA(IPG2211A2N)'] =temp['IPG2211A2N'].rolling(window=12).mean() #rolling mean mean taking 12 rows
temp['12-month-Std(IPG2211A2N)'] =temp['IPG2211A2N'].rolling(window=12).std() #rolling mean mean taking 12 rows
temp[['IPG2211A2N','12-month-SMA(IPG2211A2N)','12-month-Std(IPG2211A2N)']].plot()
```
